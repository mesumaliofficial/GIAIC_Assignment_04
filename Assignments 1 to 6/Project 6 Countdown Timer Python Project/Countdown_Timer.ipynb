# **Countdown Timer Python Project**

```python
import time

def countdown(t):
    while t:  # Jab tak t non-zero hai, loop chalega
        mins, secs = divmod(t, 60)  # divmod() se seconds ko minutes aur seconds mein divide karte hain
        timer = "{:02d}:{:02d}".format(mins, secs)  # Timer ko 2 digit format mein show karenge
        print(timer, end="\r")  # Timer print karenge aur line ko overwrite karenge
        time.sleep(1)  # 1 second ka delay
        t -= 1  # Timer ko 1 second kam karenge

    print("Timer Completed!")

t = int(input("Enter The time in Seconds: "))  # User se seconds mein time input karenge aur integer mein convert karenge
countdown(t)
```
