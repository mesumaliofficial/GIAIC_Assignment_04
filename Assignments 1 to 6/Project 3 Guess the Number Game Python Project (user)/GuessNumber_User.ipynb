# **Guess the Number Game Python Project (user)**

```python
import random

def Comp_guess(x):
  print("Welome to the guess number game")
  print(f"i am thinking of a number between 1 to {x}")

  low = 1
  high = x
  feedback = ''

  while feedback != 'c':
    if low != high:
      guess = random.randint(low, high)
    else:
      guess = low
    feedback = input(f"Is {guess} too High (H), too Low (L), or Correct (C)? ").lower()
    if feedback == "h":
      high = guess - 1
    elif feedback =="l":
      low = guess + 1
    elif feedback == "c":
      print(f"Yay! The computer guessed your number {guess} correctly!")
    else:
      print("Invalid input. Please enter H, L, or C.")

Comp_guess(10)
```
