# **Mad libs Python Project**


```python
# Mad Libs Game

print("Welcome to Mad Libs Project \n")

name: str = input(("Enter Your name: "))
place: str = input(("Enter place name: "))
f_animal: str = input(("Enter animal name: "))
l_animal: str = input (("Enter second animal name: "))


story: str = f" A {name} tricks {place} into thinking a {f_animal} is attacking his {l_animal}, \
 but when a {f_animal} actually appears, \
  no one believes him. The moral of this story is that lying can lead to a loss of trust."


print(story)
```
